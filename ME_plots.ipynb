```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_folder = r'C:\Users\cnbofan1\ABB\Safe Digital in CN - Documents - Documents\06_Test and Analytics\2022xxxx_ME_test\20220606_30s_O_30s_C'
open_num, close_num, oper_num = 0, 0, 0
for root, dirs, files in os.walk(data_folder, topdown=True):
	plt.figure(dpi=200)
	for name in files:
		if 'ang_open' in name:
			open_num += 1
			data_path = root + '\\' + name
			raw_data = pd.read_csv(data_path, header=6)
			plt.plot(raw_data * 0.01,
					 color='g',
					 linewidth='0.5')
	plt.title('open curve on 2022_06_02')
	plt.xlabel('test times with sample rate of 2500Hz')
	plt.ylabel('angle in degree')

	plt.figure(dpi=200)
	for name in files:
		if 'ang_close' in name:
			close_num += 1
			data_path = root + '\\' + name
			raw_data = pd.read_csv(data_path, header=6)
			plt.plot(raw_data * 0.01,
					 color='g',
					 linewidth='0.5')
	plt.title('close curve on 2022_06_02')
	plt.xlabel('test times with sample rate of 2500Hz')
	plt.ylabel('angle in degree')
oper_num = open_num + close_num
print('operation number', oper_num)
```
