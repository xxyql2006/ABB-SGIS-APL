# 4_Parameter calculation
## 4.1_Open current

```python
# preview
cur_val_list = ME.DataClean.curve_preview(dir_washed, 
                                         'current_open', 
                                         up_thr=1, 
                                         low_thr=0, 
                                         avg_min=0.05, 
                                         head_up_thr=0.1, 
                                         data_col_str='data')

# for each valid curve, calculate parameter
open_cur_list = []
for file in cur_val_list:

    coil_current_df = pd.read_csv(os.path.join(dir_washed, file),
                                    header=0)
    coil_current_arr = np.array(coil_current_df['data'])
    
    # calculate current parameters
    _, _, _, _, current_avg = ME.MechOperMconfig.cal_cur_avg_noself(coil_current_arr,
                                                                    invalid_title='invalid_open_current_curve {}'.format(file))

    open_cur_list.append(current_avg)

# distribution plot
print('---------------------------------------------------------------------------------------------------')
print('{:<18}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}'.format('para', 'std',
                                                                        'start', 'end',
                                                                        'max', 'mean',
                                                                        'median', 'min',
                                                                        'span', 'alm_u',
                                                                        'alm_l'))
print('---------------------------------------------------------------------------------------------------')
ME.MechOperMconfig.para_dist_plot(open_cur_list,
                                  title='open coil current',
                                  ylabel='current in amper')
```

## 4.2_Close current

```python
# preview
val_list = ME.DataClean.curve_preview(dir_washed, 
                                          'current_close', 
                                          up_thr=1.2, 
                                          low_thr=0, 
                                          avg_min=0.05, 
                                          head_up_thr=0.1, 
                                          tail_up_thr=0.1, 
                                          data_col_str='data')

# for each valid curve, calculate parameter
close_cur_list = []
for file in val_list:

    coil_current_df = pd.read_csv(os.path.join(dir_washed, file),
                                    header=0)
    coil_current_arr = np.array(coil_current_df['data'])
    
    # calculate current parameters
    _, _, _, _, current_avg = ME.MechOperMconfig.cal_cur_avg_noself(coil_current_arr)
    close_cur_list.append(current_avg)

# distribution plot
print('---------------------------------------------------------------------------------------------------')
print('{:<18}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}'.format('para', 'std',
                                                                        'start', 'end',
                                                                        'max', 'mean',
                                                                        'median', 'min',
                                                                        'span', 'alm_u',
                                                                        'alm_l'))
print('---------------------------------------------------------------------------------------------------')
ME.MechOperMconfig.para_dist_plot(close_cur_list,
                                  title='close coil current',
                                  ylabel='current in amper')
```

## 4.3_Motor current

```python
# preview
val_list = ME.DataClean.curve_preview(dir_washed, 
                                      'current_motor', 
                                      up_thr=2, 
                                      low_thr=0, 
                                      avg_min=0.1, 
                                      tail_up_thr=0.3, 
                                      data_col_str='data')

# for each valid curve, calculate parameter
motor_curr_list, charge_time_list = [], []
cnt_inval = 0
for file in val_list:

    current_df = pd.read_csv(os.path.join(dir_washed, file),
                             header=0)
    current_arr = np.array(current_df['data'])
    
    # calculate current parameters
    start_ix, end_ix, left_ix, right_ix, current_avg, charge_time = ME.MechOperMconfig.cal_motor_para_noself(current_arr,
                                                                                                            threshold=0.25)
    if (charge_time <= 1) and (cnt_inval <= 10):
        cnt_inval += 1
        print(start_ix, ';', end_ix,';', left_ix,';',right_ix)
        plt.figure()
        plt.plot(current_arr)
    elif charge_time >= 10:
        pass
    else:
        motor_curr_list.append(current_avg)
        charge_time_list.append(charge_time)

# distribution plot

print('---------------------------------------------------------------------------------------------------')
print('{:<18}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}{:<8}'.format('para', 'std',
                                                                        'start', 'end',
                                                                        'max', 'mean',
                                                                        'median', 'min',
                                                                        'span', 'alm_u',
                                                                        'alm_l'))
print('---------------------------------------------------------------------------------------------------')
ME.MechOperMconfig.para_dist_plot(motor_curr_list,
                                  title='motor current',
                                  ylabel='A')
ME.MechOperMconfig.para_dist_plot(charge_time_list,
                                  title='charge time',
                                  ylabel='S')
```
